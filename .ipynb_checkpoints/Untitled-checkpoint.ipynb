```python
import matplotlib.pyplot as plt
from numpy import *
from math import *

def linear_(x, y, x_0):
    nodes_number, i = x.size, 0
    while (abs(x_0) > abs(x[i + 1]) and i < nodes_number): i += 1
    return y[i] + (x_0 - x[i]) / (x[i + 1] - x[i]) * (y[i + 1] - y[i])

def NewTone(x, y, x_0):
    
    def f(i, j = None):
        if j is None or i == j:
            return y[i]
        else:
            k = j - i
            return (f(i + 1, i + k) - f(i, i + k - 1)) / (x[i + k] - x[i])
    
    P = 0
    for i in range(x.size):
        x_multiply = 1
        for j in range(i):
            x_multiply *= (x_0 - x[j])
        P += x_multiply * f(0, i)
    return P

def Spline(x, y, x_0):
    def h(i): return x[i] - x[i - 1]
    def u_2(i): return (y[i] - y[i - 1]) / h(i)
    def u_3(i): return (u_2(i) - u_2(i - 1)) / (x[i] - x[i - 2])
    a = y
    b = zeros(x.size)
    d = zeros(x.size)
    A = zeros((x.size, x.size))
    l = zeros(x.size)
    for i in range (1, x.size - 1):
        l[i] = 6 * u_3(i + 1)
        if i > 1: A[i][i - 1] = h(i) / (h(i) + h(i + 1))
        A[i][i] = 2
        if i < x.size - 2: A[i][i + 1] = h(i + 1) / (h(i) + h(i + 1))
    
    A_ = zeros((x.size - 2, x.size - 2))
    l_ = zeros(x.size - 2)
    for i in range (1, x.size - 1):
        for j in range(1, x.size - 1):
            A_[i - 1][j - 1] = A[i][j]
            l_[i - 1] = l[i]
    
    c = append(append(0, linalg.solve(A_, l_)), 0)
    b[1] = c[1] * h(1) / 3 + u_2(1)
    d[1] = h(1) / c[1]
    for i in range(1, x.size):
        b[i] = c[i] * h(i) / 3 + c[i - 1] * h(i) / 6 + u_2(i)
        d[i] = (c[i] - c[i - 1]) / h(i)
    
    i = 0
    while i < x.size - 1 and x[i] < x_0 : i += 1
#     return int(a[i] + b[i] * (x_0 - x[i]) + c[i] * pow((x_0 - x[i]), 2) / 2 + d[i] * pow((x_0 - x[i]), 3) / 6)
    return a[i] + b[i] * (x_0 - x[i]) + c[i] * pow((x_0 - x[i]), 2) / 2 + d[i] * pow((x_0 - x[i]), 3) / 6

def main():
    x = array([ 1910, 1920, 1930, 1940, 1950, 1960, 1970, 1980, 1990, 2000 ])
    y = array([ 92228496, 106021537, 123202624, 132164569, 151325798, 179323175, 203211926, 226545805, 248709873, 281421906 ])
    
    print('Newton extrapolation for year 2010 -', NewTone(x, y, 2010))
    print('Spline extrapolation for year 2010 -', Spline(x, y, 2010))
    print('Real USA population at year 2010 -', 308745538)
    print('Newton differnce -', NewTone(x, y, 2010) - 308745538)
    print('Spline differnce -', Spline(x, y, 2010) - 308745538)

    grid = arange(1910, 2000, 1)

    plt.plot(x, y)
    plt.plot(grid, [NewTone(x, y, grid[i]) for i in range(grid.size)])
    plt.title('newtone approx')
    plt.show()
    plt.plot(x, y)
    plt.plot(grid, [Spline(x, y, grid[i]) for i in range(grid.size)])
    plt.title('spline approx')
    plt.show()

main()
```
